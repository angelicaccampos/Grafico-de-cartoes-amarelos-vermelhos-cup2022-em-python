# Exemplo de grafico 

```python
import plotly.graph_objects as go
import plotly.express as px

colors = ['gold', 'mediumturquoise', 'darkorange', 'lightgreen']


fig = go.Figure(data=[go.Pie(
    title='Cartao amarelo, cartao vermelho e faltas - Time 1(nome do time 1) vs Time 2(nome do time 2) ',
    labels=['Cart達o Amarelo time 1','Cart達o Amarelo time 2', 'Cart達o Vermelho time 1','Cart達o Vermelho time 2','Faltas time 1','Faltas time 2'],
    values=[1000,1000,225, 225, 2250,2250])])

fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=10,
                  marker=dict(colors=colors, line=dict(color='#000000', width=1)))
fig.show()
```
